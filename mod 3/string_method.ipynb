```python
txt = "assalamualaikum Everyone! Welcome to Machine Learning Course"
processed_txt = txt.lower()
processed_txt2 = txt.upper()
processed_txt3 = txt.capitalize()

print(txt)
print(processed_txt)
print(processed_txt2)
print(processed_txt3)
```

```python
# Length #
print(len(txt))
```

```python
# Finding words in the string #
print("Machine" in txt)  #
print("machine" in txt)
print("machine" in processed_txt)
print("LEARNING" in processed_txt2)
print("learning" in processed_txt2)
```

```python
# Finding from which index the substring is starting #
start_index = txt.find("Machine") # from starting

end_index = txt.rfind("Machine") # from ending 

print(start_index)
print(end_index)
```

```python
# Counting the substring #
string = "I am Fuad. I am a Student. I am a Programmer"

count_string = string.count("I")

print(count_string)
```

```python
# Replacing a substring #
string = "I am Fuad. I am a Student. I am a Programmer"

new_string = string.replace("Programmer", "Future Programmer")

print(new_string)
```

```python
# Center align the string #
string = "I am Fuad. I am a Student. I am a Programmer"

center_string = string.center(100," ")
print(center_string)
```

```python
# finding the encode of the string #
string = "I am Fuad. I am a Student. I am a Programmer"

x = string.encode()

print(x)

print(string.encode(encoding="ascii", errors="Student"))
```

```python
# Find the starting of a string #
string = "I am Fuad. I am a Student. I am a Programmer"
x = string.startswith("I")
x_x =string.startswith("a", 2, 30)

print(x, x_x)

# Find the ending of a string #
y = string.endswith(".")
y_y = string.endswith("",43,43) # endswith(value, start, end)

print(y, y_y)
```

```python
# expanding tabs between substrings #
string = "I am Fuad. \tI am a Student. \tI am a Programmer"

x = string.expandtabs(10)

print(x)
print(string.expandtabs(30))
print(string.expandtabs(50))
print(string.expandtabs(70))
```

```python
# formatted string returns #
string = "I am {name}. I am a {profession}. I am {age} years old"
det_string = {"name" : "Fuad", "profession" : "Student", "age" : 23}

print(string.format_map(det_string))
```

```python
string = "I am Fuad. I am a Student. I am a Programmer"
string2 = "fuad1234"
string3 = "123456"
string4 = "Fuad"
string5 = "123#*wefu"
# Finding the substrings are in character or not #
print(string.isalpha())
print(string2.isalpha())
print(string3.isalpha())
print(string4.isalpha())
print(string5.isalpha())
print()

# Ascii or not #
print(string.isascii())
print(string2.isascii())
print(string3.isascii())
print(string4.isascii())
print(string5.isascii())
print()

# Decimal or not #
print(string.isdecimal())
print(string2.isdecimal())
print(string3.isdecimal())
print(string4.isdecimal())
print(string5.isdecimal())
print()

# is digit or not #
print(string.isdigit())
print(string2.isdigit())
print(string3.isdigit())
print(string4.isdigit())
print(string5.isdigit())
print()

# is a identifier or not #
print(string.isidentifier())
print(string2.isidentifier())
print(string3.isidentifier())
print(string4.isidentifier())
print(string5.isidentifier())
```

```python
string = "I am Fuad. I am a Student. I am a Programmer"
string2 = "fuad1234"
string3 = "123456"
string4 = "Fuad"
string5 = "123#*wefu"

# is lower or not #
print(string.islower())
print(string2.islower())
print(string3.islower())
print(string4.islower())
print(string5.islower())
print()

# is numeric or not #
print(string.isnumeric())
print(string2.isnumeric())
print(string3.isnumeric())
print(string4.isnumeric())
print(string5.isnumeric())
print()

# is printable or not #
print(string.isprintable())
print(string2.isprintable())
print(string3.isprintable())
print(string4.isprintable())
print(string5.isprintable())
print()
```
