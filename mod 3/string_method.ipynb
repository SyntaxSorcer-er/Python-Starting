```python
txt = "assalamualaikum Everyone! Welcome to Machine Learning Course"
processed_txt = txt.lower()
processed_txt2 = txt.upper()
processed_txt3 = txt.capitalize()

print(txt)
print(processed_txt)
print(processed_txt2)
print(processed_txt3)
```

```python
# Length #
print(len(txt))
```

```python
# Finding words in the string #
print("Machine" in txt)  #
print("machine" in txt)
print("machine" in processed_txt)
print("LEARNING" in processed_txt2)
print("learning" in processed_txt2)
```

```python
# Finding from which index the substring is starting #
start_index = txt.find("Machine") # from starting

end_index = txt.rfind("Machine") # from ending 

print(start_index)
print(end_index)
```
