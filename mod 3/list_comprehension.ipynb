```python
even = []

for i in range(101) :
    if i % 2 == 0 :
        even.append(i)

print(even)
```

list comprehension

```python
random_list = [x + 10 for x in range(1,101) if x % 2 == 0]

print(random_list)
```

```python
fruits = ['apple', 'banana', 'lichi']

capitalize_fruits = []

for fruit in fruits :
    capitalize_fruits.append(fruit.capitalize())

print(capitalize_fruits)

upper_fruits = []

for fruit in fruits :
    upper_fruits.append(fruit.upper())

print(upper_fruits)
```

```python
upper_fruits = [fruit.upper() for fruit in fruits]

print(upper_fruits)

capitalize_fruits = [fruit.capitalize() for fruit in fruits]

print(capitalize_fruits)
```
