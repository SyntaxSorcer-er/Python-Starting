Splitting

```python
prompt = "What is your name ?"
prompt2 = "what-is-your-name-?"
prompt3 = "what,is,your,name,?"

token = prompt.split()
token2 = prompt2.split("-")
token3 = prompt3.split(",")

print(token)
print(token2)
print(token3)
```

Splitting --> Joining 

```python
prompt = "What is your name ?"

token = prompt.split()

print(token)

sentence = "-".join(token)

print(sentence)
```
