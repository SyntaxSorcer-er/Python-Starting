Splitting

```python
prompt = "What is your name ?"
prompt2 = "what-is-your-name-?"
prompt3 = "what,is,your,name,?"

token = prompt.split()
token2 = prompt2.split("-")
token3 = prompt3.split(",")

print(token)
print(token2)
print(token3)
```

Splitting --> Joining 

```python
prompt = "What is your name ?"

token = prompt.split()

print(token)

sentence = "-".join(token)

print(sentence)

print(type(token))
print(type(sentence))
```

Formatted String

```python
name = "Fuad"
age = 23
height = 5.1123456
cgpa = 3.80

info = f"My name is {name.upper()}. I am {age} years old. My height is {height:.3} ft. My cgppa is {cgpa:.2}.\n"

print(info)
print(type(info))

model_accuracy = 0.876543

print(f"The model accuracy is : {model_accuracy:.2%}")
```
