```python
# converting an iterable into a string #
string = ("Yasir","Fuad","AIUB")
string2 = "Hello Everyone"
x = "#".join(string)
y = "_".join(string2)
print(x)
print(y)

myDict = {"name": "Fuad", "country": "BD"}
mySeparator = "__"

z = mySeparator.join(myDict)

print(z)
```

```python
# adding justified version of the string in the left #
txt = "I am Yasir Fuad"
x = txt.ljust(20,"_")
print(x)
```

```python
# Romoving leading characters from the left side of the string #
text = "...,,,,,1234##*`Hello"
text2 = "     Fuad     "

x = text.lstrip(".,1234#*`")
y = text2.lstrip()

print(x)
print(y)

```

```python
# Replacing characters in the string #
string = "Hello Everyone! I am Yasir Fuad"
x = str.maketrans("E", "O",) # x = str.maketrans("E", "O", "F")
y = string.translate(x)

print(y)
```

```python
# make partitions in the string #
par_string = string.partition("Everyone!")

print(par_string)
```
