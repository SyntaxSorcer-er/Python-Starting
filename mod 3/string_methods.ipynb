```python
# converting an iterable into a string #
string = ("Yasir","Fuad","AIUB")
string2 = "Hello Everyone"
x = "#".join(string)
y = "_".join(string2)
print(x)
print(y)

myDict = {"name": "Fuad", "country": "BD"}
mySeparator = "__"

z = mySeparator.join(myDict)

print(z)
```

```python
# adding justified version of the string in the left #
txt = "I am Yasir Fuad"

x = txt.ljust(20,"_")
print(x)

# adding justified version of the string in the right #
y = txt.rjust(20,"_")
print(y)
```

```python
# Removing leading characters from the string #
string1 = "...,,,*###123Yasir***`.,.,"
string2 = "     Fuad    "

print(string1.strip(".,`123#*"))
print(string2.strip())

# Romoving leading characters from the left side of the string #
text = "...,,,,,1234##*`Hello"
text2 = "     World     "
text3 = "Wow...,,,,123#*"

x = text.lstrip(".,1234#*`")
y = text2.lstrip()

print(x)
print(y)

# Romoving leading characters from the left side of the string #
p = text3.rstrip(".,`#*123")
q = text2.rstrip()

print(p)
print(q)
```

```python
# Replacing characters in the string #
string = "Hello Everyone! Everyone! I am Yasir Fuad"
x = str.maketrans("E", "O",) # x = str.maketrans("E", "O", "F")
y = string.translate(x)

print(y)
```

```python
# make partitions in the string #
par_string = string.partition("Everyone!")

print(par_string)

# make reverse partitions in the string #
par_string2 = string.rpartition("Everyone!")

print(par_string2)
```

```python
# replacing substring with new substring #
rep_string = string.replace("Hello", "Assalamualaikum",)

print(rep_string)

string2 = "Hello Class Class. I am Fuad"
rep_string2 = string2.replace("Class", "All", 2)

print(rep_string2)
```

```python
# splitting #
x = string2.split() # split(separator, maxsplit)
print(x)

# reverse splitting #
y = string2.rsplit(" ",2) # rsplit(separator,maxsplit)
print(y)

# split at line breaks #
texts = "Hello I am Fuad\nI am a student"

print(texts.splitlines())
```

```python
# Swapping the case of the string #
string2 = "Hello Class Class. I am Fuad"
print(string2.swapcase())
```

```python
# converting 1st letter of every substring into uppercase #
my_text = "hello everyone i am fuad"
print(my_text.title())
```
