```python
n = int(input())
inp = input()
numbers = inp.split()

mini = int(numbers[0])
maxi = int(numbers[0])

min_index = 0
max_index = 0


for i in range(n):
    x = int(numbers[i])

    if x < mini :
        mini = x
        min_index = i

    if x > maxi :
        maxi = x 
        max_index = i     

numbers[min_index], numbers[max_index] = numbers[max_index], numbers[min_index]

for num in numbers :
    print(num, end=" ")
```
