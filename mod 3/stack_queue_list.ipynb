```python
# small to large #
lists = [20,70,10,50,90,30,60,80,40]

lists.sort()

print(lists)

# large to small #
lists.sort(reverse=True)

print(lists)
```

Stack as list --> last in first out


```python
stack = []

# stack push #
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

print(stack)

# getting the top element #
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())

print(stack)
```
