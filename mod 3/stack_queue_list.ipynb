```python
# small to large #
lists = [20,70,10,50,90,30,60,80,40]

lists.sort()

print(lists)

# large to small #
lists.sort(reverse=True)

print(lists)
```

List as Stack --> last in first out


```python
stack = []

# stack push #
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)

print(f"Stack List :{stack}\n")

# getting the top element #
print(f"top element is : {stack[-1]}")
print(f"{stack.pop()} is removed")
print(f"New Stack Elements : {stack}")

print(f"top element is : {stack[-1]}")
print(f"{stack.pop()} is removed")
print(f"New Stack Elements : {stack}")

print(f"top element is : {stack[-1]}")
print(f"{stack.pop()} is removed")
print(f"New Stack Elements : {stack}")

print(f"top element is : {stack[-1]}")
print(f"{stack.pop()} is removed")
print(f"New Stack Elements : {stack}")

print(f"top element is : {stack[-1]}")
print(f"{stack.pop()} is removed")
print(f"New Stack Elements : {stack}")
```

List as Queue --> first in first out

```python
queue = []

# Queue element insertion #
queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

print(queue)


# remove #
print(f"{queue.pop(0)} is removed")

# accessing the first element #
print(f"Now front is : {queue[0]}" )

print(f"{queue.pop(0)} is removed")

print(f"Now front is : {queue[0]}" )

print(f"{queue.pop(0)} is removed")

print(f"Now front is : {queue[0]}" )

print(f"{queue.pop(0)} is removed")

print(f"Now front is : {queue[0]}" )
```
