Tuple: 


--> Collection of Data

--> Immutable

```python
# Declaration #
tup = (10, 20, 30, 40)

print(type(tup), tup)

float_tup = (10.20, 20.30, 30.40, 40.50)

print(type(float_tup), float_tup)

mixed_tup = (10, 30.50, "Fuad", True)

print(type(mixed_tup), mixed_tup)
```

```python
tup1 = (10)
tup2 = (40,)
tup3 = tuple([10, 20, 30])

print(type(tup1), tup1)
print(type(tup2), tup2)
print(type(tup3), tup3)
```

```python
lst = [60, 70, 80, 90]

tup4 = tuple(lst)

print(type(tup4), tup4)
```
