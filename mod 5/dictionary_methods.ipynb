```python
# printing default values #
dic = { "Name" : "Fuad" , "Age" : 23 , "CGPA" : 3.8 , "Numbers" : [10,20,30]}

print(dic.get("Address"))
print(dic.get("Address", 0))
```

Adding values

```python
dic = { "Name" : "Fuad" , "Age" : 23 , "CGPA" : 3.8 , "Numbers" : [10,20,30]}

print(dic)

# method 1
dic["Address"] = "Bashundhara"
print(dic)

# method 2
dic.update({"Phone No." : 8801309883852})
print(dic)
```

Deleting values

```python
del dic["Numbers"]

print(dic)
```

Copying dictionaries

```python
dic_2 = dic.copy()

print(dic_2)
```

```python
# Testing #
dic_3 = {(10, 20, 30) : [10, 20, 30]} # tuple as a key as it is immutable
print(dic_3)

dic_4 = {"key" : {"name" : "fuad", "age" : 23}}
print(dic_4)
```

Iterating over Dictionary

```python
dic = { "Name" : "Fuad" , "Age" : 23 , "CGPA" : 3.8 , "Numbers" : [10,20,30]}

keys = dic.keys()

print(keys)

dic["Address"] = "Bashundhara"

print(keys)

values = dic.values()

print(values)

dic["Age"] = 24

print(values)

items = dic.items()

print(items)
```

```python
# iteration #

for key,value in dic.items() :
    print(key,":",value)
```
