Dictionaries (key, value)

--> mutable

--> unordered 

```python
dic = {}
print(type(dic))

dic = { "Name" : "Fuad" , "Age" : 23 , "CGPA" : 3.8 , "Numbers" : [10,20,30]}
print(type(dic), dic)
```

Access methods

```python

# method 1 #
print(dic["Name"]) # dic_name[key]
print(dic["Age"])
print(dic["CGPA"])
print(dic["Numbers"])

# method 2 #
print(dic.get("Name"))
print(dic.get("Age"))
print(dic.get("CGPA"))
print(dic.get("Numbers"))
```

New value assign 

```python
dic = { "Name" : "Fuad" , "Age" : 23 , "CGPA" : 3.8 , "Numbers" : [10,20,30]}

dic["Name"] = "Yasir Faud"

print(dic)

dic = { "Name" : "Fuad" , "Age" : 23 , "CGPA" : 3.8 , "Numbers" : [10,20,30], "Age" : 30 }
print(dic) # always replace old values with latest values 
 
```
