1. Given a list of numbers, make a list with unique values.

```python
lst = [10, 20, 10, 30, 30, 50, 30, 10, 20, 10, 10]

unique_values = set(lst) # list to set conversion

# print(unique_values)

lst = list(unique_values) # set to list conversion

print(lst)
```

2. Given a string, print the frequency of the words

```python
string = """I am a motivated and responsible individual with hands-on experience in volunteering and event organization. I worked as a volunteer at the 2nd Space Exploration Olympiad held at AIUB and served as an organizing team member at the Global Scholars Olympiad held at IUB, both organized by Space Innovation Camp. These experiences helped me develop strong teamwork, communication, and event management skills. I am eager to contribute, learn, and support the success of this AI-focused event."""

words = string.split()

count = {}

for word in words :
    count[word] = count.get(word, 0) + 1

for key,value in count.items() :
    print(f"count of {key} is:{value}")
```
