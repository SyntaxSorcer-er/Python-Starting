Problem name: Sensor Calibrator — Min-Max Scaling

You’re calibrating a tiny sensor before training a model. 

Given x, min_v, and max_v, compute:

norm = (x - min_v) / (max_v - min_v)

Then print norm with 2 decimal places.

New concept — Feature scaling

Models learn better when inputs share similar scales;
 
min–max transform values into [0, 1] using a equation

Input Format

One line with three real numbers:

x min_v max_v

Constraints

Assume max_v > min_v

Basics only: input/output and arithmetic

No lists/arrays; no loops required

Output Format

Print a single real number with 2 decimal places (e.g., 0.42).

Sample Input 0

0 0 10

Sample Output 0

0.00

Sample Input 1

2.5 0 10

Sample Output 1

0.25

```python
inp = input().split()

x = float(inp[0])
min_v = float(inp[1])
max_v = float(inp[2])

norm = (x - min_v) / (max_v - min_v)

print(f"{norm:.2f}")
```
