```python
s = { 5, 1, 3, 2, 4 }
print(s , "\n")
# set in unordered so indexing not possible 

s.add(10) 
s.add(0)

print(s, "\n")

s.pop() # deletes 0
s.pop() # deletess 1

print(s, "\n")

s.remove(10)

print(s, "\n")

s.clear() #removes all elements

print(s)
```

```python
a = { 1, 2, 3 }
b = { 1, 2, 3, 4, 5, 6 }
c = { 4, 5, 6 }

print("a & b all elements: ", a.union(b))

print("a & b common elements: ", a.intersection(b))

print("a is a disjoint of b :", a.isdisjoint(b)) # check if there are any common elements between 2 sets. if yes it will return False otherwise True

print("a is a disjoint of c :", a.isdisjoint(c))

print("a is a subset of b :", a.issubset(b))

print("a is a subset of c :", a.issubset(c))

print("c is a subset of b :", c.issubset(b))
```

```python
a = { 1, 2, 3 }
b = { 1, 2, 3, 4, 5, 6 }
c = { 4, 5, 6 }

print("a - b:", a.difference(b))
print("b - a:", b.difference(a))
print("a - c:", a.difference(c))
print("c - a:", c.difference(a))
print("b - c:", b.difference(c))
print("c - b:", c.difference(b))

d = {"apple", "banana", "cherry"}
e = {"google", "microsoft", "apple"}
f = {"cherry", "micra", "bluebird"}

print("d - e - f:", d.difference(e,f))
print("e - d - f:", e.difference(d,f))
print("f - d - e", f.difference(d,e))

# always returs 1st set elements 

print("d - e - f:", d - e - f)
```
