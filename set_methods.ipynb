```python
s = { 5, 1, 3, 2, 4 }
print(s , "\n")
# set in unordered so indexing not possible 

s.add(10) 
s.add(0)

print(s, "\n")

s.pop() # deletes 0
s.pop() # deletess 1

print(s, "\n")

s.remove(10)

print(s, "\n")
```

mathematical methods 


```python
a = { 1, 2, 3 }
b = { 2, 3, 4, 5, 6 }

print(a.union(b))

print(a.intersection(b))

print(a.isdisjoint(b)) # check if there are any common elements between 2 sets. if yes it will return False otherwise True
```
