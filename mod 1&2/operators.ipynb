Mathematical Operators

```python
x = 10
y = 3

sum = x + y
sub = x - y
mult = x * y
div = x / y
rem = x % y
print(sum, sub, mult, div, rem)
```

```python
x = 10 ; y = 3

sum = x + y
sub = x - y
mult = x * y
div = round(x / y , 2)
rem = x % y
print(sum, sub, mult, div, rem)
```

Comparison Operators

```python
x = 10
y = 3

greater_than = x > y
greater_than_equal = (10 >= 10)
print(greater_than , greater_than_equal)

less_than = x < y
less_than_equal = (10 <= 10)
print(less_than , less_than_equal)

equal = x == y
print(equal)

not_equal = x != y
print(not_equal)
 
```

Logical Operator

```python
x = 10; y = 5; z = 3

result = (x > y) and (x > z)

print(result)
```
