```python
n = int(input())
inp = input()
numbers = inp.split()

positive = 0
negative = 0
even = 0 
odd = 0

for i in range(n):
    x = int(numbers[i])

    #for positive and negative
    if x > 0 :
        positive += 1
    elif x < 0 : 
        negative += 1
    #for even and odd
    if x % 2 == 0 :
        even += 1
    else :
        odd += 1

print("Even :" , even)
print("odd :" , odd)
print("Positive :" , positive)
print("Negative :" , negative)
```
