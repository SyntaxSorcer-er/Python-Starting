Types of Machine Learning :
  1. supervised
  2. unsupervised
  3. reinforcement

1. Supervised Learning

This type of learning uses labeled data, meaning each input comes with a correct output.

> The model learns a mapping: input → output

> It tries to minimize the difference between predicted and actual results

Common tasks:

> Classification (e.g., spam vs non-spam emails)

> Regression (e.g., predicting house prices)

Example:
You train a model with images of cats and dogs (labeled), and it learns to classify new images correctly.

2. Unsupervised Learning

This works with unlabeled data, meaning there are no predefined outputs.

> The model finds hidden patterns or structures in the data

> No “correct answer” is given during training

Common tasks:

> Clustering (grouping similar data points)

> Dimensionality reduction (simplifying data)

Example:
Grouping customers based on purchasing behavior without knowing categories beforehand.

3. Reinforcement Learning

This is based on learning through interaction and feedback.

> An agent takes actions in an environment

> It receives rewards or penalties

> Goal: maximize total reward over time

Key idea:

<> Trial-and-error learning with delayed feedback

Example:
Training a game-playing AI (like chess or video games) where it improves by winning or losing.

Quick Comparison

Supervised → learns from labeled data

Unsupervised → finds patterns in unlabeled data

Reinforcement → learns from rewards and actions
