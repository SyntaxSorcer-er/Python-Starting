Set is an inbuild data structure in python

--> stores unique elements

--> mutable

--> unordered (no indexing, no slicing)


```python
A = {1, 2, 3}

print(type(A), A)
```

```python
# empty structures #

B = {}  # dictionary type
C = []  # list type
D = ()  # tuple type

print(type(B), B)
print(type(C), C)
print(type(D), D)
print()

S = set()
print(type(S), S)
```

Set access

```python
S = { 1, 2, 3, 4, 1, 2, 3 }
print(S, "\n")

if (3 in S) :
    print("Yes, 3 is present in set S \n")
else :
    print("No, 3 is not present in set S \n")

for elements in S :
    print(elements)

sum = 0 
for elements in S :
    sum += elements

print("\nSum is :", sum)
```
