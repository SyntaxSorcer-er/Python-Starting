Set is an inbuild data structure in python

--> stores unique elements

--> mutable

--> unordered (no indexing, no slicing)


```python
A = {1, 2, 3}

print(type(A), A)
```

```python
B = {}  # dictionary type
C = []  # list type
D = ()  # tuple type

print(type(B), B)
print(type(C), C)
print(type(D), D)
```
